```python
import os
import warnings

import pandas as pd
import numpy as np
pd.set_option("display.max_columns", 50)
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
from pyhive import hive
from geopy import distance
```

```python
warnings.simplefilter(action='ignore', category=UserWarning)

username = os.environ['RENKU_USERNAME']
hiveaddr = os.environ['HIVE_SERVER2']
(hivehost, hiveport) = hiveaddr.split(':')

# create connection
conn = hive.connect(host=hivehost, 
                    port=hiveport,
                    username=username) 
# create cursor
cur = conn.cursor()
```

```python
# create your database if it does not exist
query = """
CREATE DATABASE IF NOT EXISTS {0} LOCATION '/user/{0}/hive'
""".format(username)
cur.execute(query)

# make your database the default
query = """
USE {0}
""".format(username)
cur.execute(query)
```

```python
# drop table if exists
query = """
DROP TABLE IF EXISTS {0}.stop_times
""".format(username)
cur.execute(query)
```

```python
# create table
query = """
    CREATE EXTERNAL TABLE {0}.stop_times(
        trip_id STRING,
        arr_time STRING,
        dep_time STRING,
        stop_id STRING,
        stop_sequence STRING,
        pickup_type STRING,
        drop_off_type STRING
    )
    STORED AS ORC
    LOCATION '/data/sbb/part_orc/stop_times/year=2019/month=05/day=08'
    TBLPROPERTIES ("orc.compress"="SNAPPY")
""".format(username)
cur.execute(query)

# repair table if needed
query = """MSCK REPAIR TABLE {0}.stop_times""".format(username);
cur.execute(query)
```

```python

# trip_id, arr_time, dep_time, stop_id, stop_sequence 
query = """
    SELECT *
    FROM {0}.stop_times
    WHERE stop_id = 8500926
    LIMIT 1000
""".format(username)
stop_times = pd.read_sql(query, conn)
```
